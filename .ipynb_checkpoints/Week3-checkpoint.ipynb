##Make a series with cumulative miles ridden
I realized I did this wrong - In this example I gave daily distances and 
made them cumulative instead of the other way around. 
In the second panel that has been fixed.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

distance = pd.Series([55,120,77,49,63,38,66,52,85,107]).cumsum()
distance.index += 1
distance.columns = ['Day', 'Cumulative Distance']

plt.plot(distance, linestyle='--', marker='o')
plt.ylabel('Miles')
plt.xlabel('Day')
plt.suptitle('Cumulative Miles Ridden')
plt.grid()
plt.show()

print(distance.columns)
print(distance)
print('Total miles biked: ',distance.values[-1])
```

##Below are the cumulative distances that are then broken down
##into daily distances

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

cumdist = pd.Series([55,120,198,227,283,311,367,399,425,459])
cumdist.index += 1

day1 = pd.Series(cumdist.iloc[0])
day1.index += 1

otherdays = pd.Series(np.diff(cumdist))
otherdays.index += 2

alldays = [day1, otherdays]
dailydist = pd.concat(alldays)

plt.plot(dailydist, linestyle='--', marker='o')
plt.ylabel('Miles')
plt.xlabel('Day')
plt.suptitle('Daily Miles Ridden')
plt.grid()
plt.show()
```
